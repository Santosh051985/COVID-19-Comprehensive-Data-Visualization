# Data Visualization

I have tried to use different types of plotly map visualizations to show examples of it.

***A. Top 20 Active Case  Analysis:***

```python
top_actives = df_1.groupby(by = 'State')['Active'].sum().sort_values(ascending = False).head(15).reset_index()
top_actives
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize= (15,10))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Total cases",fontsize = 30)
plt.ylabel('Country',fontsize = 30)
plt.title("Top 20 state having most active cases" , fontsize = 30)
ax = sns.barplot(x = top_actives.Active, y = top_actives.State)
for i, (value, name) in enumerate(zip(top_actives.Active, top_actives.State)):
    ax.text(value, i-.05, f'{value:,.0f}',  size=10, ha='left',  va='center')
ax.set(xlabel='Total cases', ylabel='State')
```

****Here are my observations :***
- As the covid-19 testing is increasing, The active number of cases is also increasing day by day.
-  The number of active cases is on a high in the Maharashtra, Delhi, TamilNadu, and Rajasthan.
